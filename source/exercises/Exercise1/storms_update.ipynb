```python
import pandas as pd
import geopandas as gpd

df1 = pd.read_csv('Year.2015.ibtracs_wmo.v03r10.csv', sep=',', encoding='latin1')
df2 = pd.read_csv('Year.2016.ibtracs_wmo.v03r10.csv', sep=',', encoding='latin1')
df3 = pd.read_csv('Year.2017.ibtracs_wmo.v03r10.csv', sep=',', encoding='latin1')

pd.set_option('max_columns',20)

df = pd.concat([df1, df2, df3])

display(df.sample(50))
```

```python
grouped = df.groupby('Serial_Num')
len(grouped)

news = []

for name, group in grouped:
    #print(f'name: {name}')
    g = group.sort_values(["year"])
    first = g.iloc[0]
    last = g.iloc[-1]
    news.append( {  "Name": first["Name"],
                    "Serial_Num": first["Serial_Num"],
                    "year": first['year'],
                    "Basin": first['Basin'],
                    "Sub_basin": first['Sub_basin'],
                    "Num": first['Num'],
                    "Latitude_first": first['Latitude'],
                    "Longitude_first": first['Longitude'], 
                    "Latitude_last": last['Latitude'],
                    "Longitude_last": last['Longitude'],
                    "ISO_time_first": first['ISO_time'],
                    "ISO_time_last": last['ISO_time'],
                    "Nature": first['Nature']})

storms = pd.DataFrame(news)
storms.to_csv('Years.2015-2017.ibtracs_wmo.storms.csv', sep=',', encoding='latin1', index=False)
```
